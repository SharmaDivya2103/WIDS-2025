```python
import json
import pandas as pd

with open("polygon_news_sample.json", "r", encoding="utf-8") as f:
    data = json.load(f)

type(data)
```

```python
data[0]
```

```python
aapl_articles = []

for item in data:
    if "AAPL" in item.get("tickers", []):
        aapl_articles.append(item)

len(aapl_articles)
```

```python
news_list = []

for one_article in data:
    tickers_list = one_article["tickers"]
    
    if "AAPL" in tickers_list:
        article_title = one_article["title"]
        article_date = one_article["published_utc"]
        
        insight_list = one_article["insights"]
        
        if len(insight_list) > 0:
            article_sentiment = insight_list[0]["sentiment"]
        else:
            article_sentiment = None
        
        row_data = {
            "Date": article_date,
            "Headline": article_title,
            "Sentiment": article_sentiment
        }
        
        news_list.append(row_data)

news_df = pd.DataFrame(news_list)

news_df.head()
```

```python
news_df["Date"] = pd.to_datetime(news_df["Date"])

news_df["Sentiment"] = news_df["Sentiment"].map({
    "positive": 1,
    "neutral": 0,
    "negative": -1
})

news_df = news_df.dropna()

news_df.head()
```

```python
daily_sentiment = news_df.groupby(news_df["Date"].dt.date)["Sentiment"].mean().reset_index()
daily_sentiment.columns = ["Date", "Avg_Sentiment"]

daily_sentiment.head()
```

```python
stock["Date"] = pd.to_datetime(stock["Date"]).dt.date
```

```python
stock.columns = stock.columns.get_level_values(0)
stock.head()
```

```python
stock["Date"] = pd.to_datetime(stock["Date"]).dt.date
```

```python
df = pd.merge(stock, daily_sentiment, on="Date", how="inner")
df.head()
```

```python
df.shape
```

```python
df["Return"] = df["Close"].pct_change()
df["Target"] = (df["Return"] > 0).astype(int)
df = df.dropna()

df[["Date", "Close", "Return", "Target"]].head()
```

```python
features = ["Open", "High", "Low", "Close", "Volume", "Avg_Sentiment"]
X = df[features]
y = df["Target"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
split = int(len(df) * 0.8)

X_train = X_scaled[:split]
X_test = X_scaled[split:]

y_train = y[:split]
y_test = y[split:]

len(X_train), len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)

lr_preds = lr.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

rf_preds = rf.predict(X_test)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(eval_metric="logloss")
xgb.fit(X_train, y_train)

xgb_preds = xgb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate(y_true, y_pred):
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred),
        "Recall": recall_score(y_true, y_pred),
        "F1": f1_score(y_true, y_pred)
    }

lr_metrics = evaluate(y_test, lr_preds)
rf_metrics = evaluate(y_test, rf_preds)
xgb_metrics = evaluate(y_test, xgb_preds)

results = pd.DataFrame(
    [lr_metrics, rf_metrics, xgb_metrics],
    index=["Logistic Regression", "Random Forest", "XGBoost"]
)

results
```
